```python
df_iron = df.drop(columns="% Silica Concentrate")
target = df['% Silica Concentrate']
```

```python
df_iron.shape
```

```python
import numpy as np

# Convert DataFrame to NumPy array for machine learning algorithms
df_iron = np.array(df_iron)
target = np.array(target)
```

```python
target = target.reshape(-1,1)
target.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler_x = StandardScaler()
X = scaler_x.fit_transform(df_iron)

scaler_y = StandardScaler()
y = scaler_y.fit_transform(target)

```

```python
# Split the data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import r2_score, root_mean_squared_error, accuracy_score

ridge_model = RidgeCV(alphas=(0.1, 1.0, 10.0), fit_intercept=True, cv=5, scoring='neg_root_mean_squared_error')
ridge_model.fit(X_train, y_train)

y_pred_train = ridge_model.predict(X_train)
y_pred_test = ridge_model.predict(X_test)

# R-squared (R2) score for training set
r2_train = r2_score(y_train, y_pred_train)

# Root Mean Squared Error (RMSE) for training set
rmse_train = root_mean_squared_error(y_train, y_pred_train)




print("R-squared (R2) score for Training Set:", r2_train)
print("Root Mean Squared Error (RMSE) for Training Set:", rmse_train)


print("#"*100)
#------------------------------------------------------------------------

# R-squared (R2) score
r2 = r2_score(y_test, y_pred_test)

# Root Mean Squared Error (RMSE)
rmse = root_mean_squared_error(y_test, y_pred_test)


print("R-squared (R2) score for Test Set:", r2)
print("Root Mean Squared Error (RMSE) for Test Set:", rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor()

dt_model.fit(X_train, y_train)

```

```python
acuracy = dt_model.score(X_test, y_test)

acuracy
```
